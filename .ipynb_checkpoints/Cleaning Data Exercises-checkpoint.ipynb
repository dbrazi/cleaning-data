## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
# Plot a heatmap of any missing data


cols = women_df.columns
colours = ['#6f4369', '#f28f93']
sns.heatmap(women_df[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
# Try out a missing data percentage list! 

for col in women_df.columns:
    pct_missing = np.mean(women_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Plot a missing data histogram

for col in women_df.columns:
    missing = women_df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('Missing Data: {}'.format(col))
        women_df['{} MISSING'.format(col)] = missing


ismissing_cols = [col for col in women_df.columns if 'MISSING' in col]
women_df['num_missing'] = women_df[ismissing_cols].sum(axis=1)

women_df['num_missing'].value_counts().reset_index().sort_values(by='index').plot.hist(x='index', y='num_missing')

```

Now that you have tried all 3 techniques for detecting missing data, did you find any? Reading through the article, which solution do you think will help you clean the data set and handle the missing data? Try out your solution below!

```python
# Handle any missing data in your dataframe.

women_df['Title'] = women_df['Title'].fillna('MISSING')

women_df['Review Text'] = women_df['Review Text'].fillna('MISSING')

```
